```python
import matplotlib.pyplot as plt
import numpy as np
import torch

model = torch.load('model0508.ckpt', map_location=torch.device('cpu'))
```

```python
train_loss = model['train_loss']
val_loss = model['val_loss']
plt.plot(train_loss)
plt.plot(val_loss)
plt.legend(["Train loss", "Validation Loss"], prop={'size': 12})
plt.xlim(0, min(len(train_loss), len(val_loss)))
plt.ylim(0, 2.1)
plt.xlabel('epochs', fontsize=16)
plt.ylabel('Loss', fontsize=16)
plt.savefig('loss0508-high-lvl-codes.png')
```

```python
model = torch.load('model-val.ckpt', map_location=torch.device('cpu'))
train_loss = model['train_loss']
val_loss = model['val_loss']
plt.plot(train_loss)
plt.plot(val_loss)
plt.legend(["Train loss", "Validation Loss"], prop={'size': 12})
plt.xlim(0, min(len(train_loss), len(val_loss)))
# plt.ylim(0, 2.1)
plt.xlabel('epochs', fontsize=16)
plt.ylabel('Loss', fontsize=16)
plt.savefig('loss-low-lvl-codes.png')
```

```python
train_f1 = model['train_f1']
val_f1 = model['val_f1']
plt.plot(train_f1)
plt.plot(val_f1)
plt.legend(["Train F1-micro", "Validation F1-micro"], prop={'size': 12})
plt.xlim(0, min(len(train_loss), len(val_loss)))
plt.ylim(0.35, 1)
plt.xlabel('epochs', fontsize=16)
plt.ylabel('F1', fontsize=16)
plt.savefig('f10508-high-lvl-codes.png')
```

```python
val_f1 = model['val_f1']
val_f1_macro = model['val_f1_macro']
val_precision = model['val_precision']
val_recall = model['val_recall']
plt.plot(val_f1)
plt.plot(val_f1_macro)
plt.plot(val_precision)
plt.plot(val_recall)
plt.legend(["F1-micro", 'F1-macro', 'Precision', 'Recall'], prop={'size': 12})
plt.xlim(0, len(val_f1))
plt.xlabel('epochs', fontsize=16)
plt.ylabel('Score', fontsize=16)
plt.savefig('val0508-high-lvl-codes.png')
```
